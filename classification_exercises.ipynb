```python
# Encode the categorical columns.
# Create dummy variables of the categorical columns and concatenate them onto the dataframe.

dummy_telco_data = pd.get_dummies(telco_data[['gender', 'partner', 'dependents', 'phone_service', 'multiple_lines', 'online_security', 'online_backup','device_protection','tech_support','streaming_tv','streaming_movies', 'paperless_billing', 'churn']],dummy_na=False, drop_first=[True, True])
dummy_telco_data
```

```python
# Create a function named prep_telco that accepts the raw telco data,
# and returns the data with the transformations above applied.
def clean_telco_data(telco):

    # Drop duplicates
    telco.drop_duplicates(inplace=True)
    # Drop columns 
    telco_drop = ['customer_id','payment_type','internet_service_type','contract_type']
    telco_data = telco.drop(columns = telco_drop)
    # encoded categorical variables
    dummy_telco_data = pd.get_dummies(telco_data[['gender', 'partner', 'dependents', 'phone_service', 'multiple_lines', 'online_security', 'online_backup','device_protection','tech_support','streaming_tv','streaming_movies', 'paperless_billing', 'churn']],dummy_na=False, drop_first=[True, True])

    telco = pd.concat([telco, dummy_telco_data], axis=1)
    return telco.drop(columns=['gender', 'partner', 'dependents', 'phone_service', 'multiple_lines', 'online_security', 'online_backup','device_protection','tech_support','streaming_tv','streaming_movies', 'paperless_billing', 'churn'])

prep_telco = clean_telco_data(telco)
prep_telco
```
